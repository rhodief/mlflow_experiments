# Build

## IngestÃ£o de dados

```python
df = pd.read_csv('./students.csv')
```

```python
df.head()
```

```python
df[df['Study Environment'].isna()]
```

```python
df['Study Environment'].fillna('disrupted', inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_study_env = LabelEncoder()
```

```python
df['Study Environment Numerico'] = le_study_env.fit_transform(df['Study Environment'])
```

```python
df['Study Environment Numerico'].value_counts()
```

```python
le_copy_strat = LabelEncoder()
df['What coping strategy you use as a student? Numerico'] = le_copy_strat.fit_transform(df['What coping strategy you use as a student?'])
df['What coping strategy you use as a student? Numerico'].value_counts()
```

```python
le_bad_habits = LabelEncoder()
df['Do you have any bad habits like smoking, drinking on a daily basis? Numerico'] = le_bad_habits.fit_transform(df['Do you have any bad habits like smoking, drinking on a daily basis?'])
df['Do you have any bad habits like smoking, drinking on a daily basis?'].value_counts()
```

## Criando RÃ³tulo

```python
le_study_env.classes_
```

```python
label_alvo = le_study_env.transform(['Peaceful'])[0]
df['class'] = df['Study Environment Numerico'].apply(lambda x: 0 if x == label_alvo else 1)
```

```python
df['class'].value_counts() / len(df)
```

## Treinamento

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
colunas_catetoricas = df.select_dtypes(include=["object", "category"]).columns
```

```python
caracteristicas = df.drop(columns=[*colunas_catetoricas] + ['class'])
```

```python
caracteristicas
```

```python
X = caracteristicas
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=73)
```

```python
X_train
```

```python
mlflow.set_experiment('atividade01')
mlflow.sklearn.autolog()
```

```python
modelo = DecisionTreeClassifier()
modelo.fit(X_train, y_train)
```

## Teste

```python
modelo.score(X_test, y_test)
```

```python
print(classification_report(y_test, modelo.predict(X_test)))
```

## Empacotamento

```python
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType

# Especifico que as features sÃ£o do tipo Tensor de float, seguindo a dimensÃ£o do dataset de treino
tipo_inicial = [('float_input', FloatTensorType(X_train.shape))]

with open('modelo_atividade01.onnx', 'wb') as arquivo:
    modelo_serializado = convert_sklearn(modelo, initial_types=tipo_inicial)

    arquivo.write(modelo_serializado.SerializeToString())
```

## Registro do modelo

```python
from mlflow.models import infer_signature

with mlflow.start_run(run_name='teste_foo') as run:
    assinatura = infer_signature(X_train, modelo.predict(X_test))

    mlflow.sklearn.log_model(
        sk_model=modelo,
        signature=assinatura,
        input_example=X_train,
        registered_model_name="atividade01",
        name="arvore_decisao"
    )
```
